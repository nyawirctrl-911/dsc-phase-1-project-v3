## DATA CLEANING

```python
used_columns = df[['Make','Model','Injury.Severity','Aircraft.damage','Aircraft.Category',
                   'Number.of.Engines','Publication.Date','Engine.Type','Purpose.of.flight',
                   'Weather.Condition','Total.Fatal.Injuries',
                   'Total.Serious.Injuries','Total.Minor.Injuries','Registration.Number','Investigation.Type']]

#It organizes the column titles by removing spaces, converting to lowercase and replacing periods with underscores
used_columns.columns = (
    used_columns.columns
    .str.strip()
    .str.lower()
    .str.replace('.', '_', regex=False)
)
used_columns
```

```python
used_columns.isna().sum() #counts no. of null values
```

```python
used_columns=used_columns.dropna() #removes null values
used_columns
```

```python
#only has aircraft_category as airplane 

airplane_dataset=used_columns.loc[used_columns['aircraft_category']=='Airplane']
airplane_dataset
```

# DATA MANIPULATION


```python
#Standardize the 'make' column by stripping whitespace and converting to uppercase
used_columns['make'] = (
    used_columns['make']
    .str.strip()
    .str.upper()
)

used_columns
```

```python
#make date to be for the past 20 years only
used_columns['publication_date'] = pd.to_datetime(used_columns['publication_date'], errors='coerce')
used_columns = used_columns[used_columns['publication_date'] >= '2004-01-01']
used_columns
```

```python
#Finding data that only has investigation type as Accident
accident_data=airplane_dataset.loc[airplane_dataset['investigation_type']=='Accident']
accident_data
```

```python
#Shows how often an accident for each make
accident_frequency = used_columns.groupby(['make', 'model']).size().reset_index(name='accident_count')
accident_frequency
```

```python
#makes with most accidents
top_10_make = (
    used_columns['make']
    .value_counts()
    .head(10)
)
top_10_make
```

```python
#model of aircraft with most accidents
make_model_accidents = (
    used_columns
    .groupby(['make', 'model'])
    .size()
    .reset_index(name='Total_Accidents')
    .sort_values('Total_Accidents', ascending=False)
)

make_model_accidents.head(10)
```

```python
#level of fatality after accident based on make 
fatality_analysis = (
    used_columns
    .groupby(['make', 'model'])
    .agg({
        'total_fatal_injuries': 'sum',
        'total_serious_injuries': 'sum',
        'total_minor_injuries': 'sum'
    })
    .reset_index()
    .sort_values('total_fatal_injuries', ascending=False)
)
fatality_analysis.head(10)
```

```python
#aicraft_damage after accident on the bottom 10 makes
damage_analysis = (
    used_columns
    .groupby(['make', 'aircraft_damage'])
    .size()
    .reset_index(name='Total_Accidents')
    .sort_values('Total_Accidents', ascending=False)
)
damage_analysis.head(10)
```

```python
#makes  with least accidents
bottom_10_make = (
    used_columns['make']
    .value_counts()
    .tail(10)
)
bottom_10_make
```

```python
#model with most accidents
make_model_accidents.tail(10)
```

```python
#aicraft_damage after accident on the bottom 10 makes
damage_analysis = (
    used_columns
    .groupby(['make', 'aircraft_damage'])
    .size()
    .reset_index(name='Total_Accidents')
    .sort_values('Total_Accidents', ascending=False)
)
damage_analysis.tail(10)
```

```python
#level of fatality after accident based on make 
fatality_analysis = (
    used_columns
    .groupby(['make', 'model'])
    .agg({
        'total_fatal_injuries': 'sum',
        'total_serious_injuries': 'sum',
        'total_minor_injuries': 'sum'
    })
    .reset_index()
    .sort_values('total_fatal_injuries', ascending=False)
)
fatality_analysis.tail(10)
```

- From the above date, we conclude that CESSNA and it's models are the ones leading in flight accidents for the past 20 years of which most of the accidents had fatal injures. However, a good number of planes were substantial.

- Foster also has the least accidents having 1 accident in the past 20 years hence being the best. The condition of all airplanes after accident were substantial apart from HUGHES which was destroyed. No one was injured during the accidents.

```python
#find most frequent weather condition during accidents
weather_conditions = (
    used_columns
    .groupby('weather_condition')
    .size()
    .reset_index(name='Total_Accidents')
    .sort_values('Total_Accidents', ascending=False)
)

weather_conditions.tail(10)
```

# VISUALIZING THE DATA

```python
# Plotting a bar graph for top 10 makes with most accidents

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
x=make_model_accidents.head(10)
y= x['Total_Accidents']
labels = x['make']
plt.bar(labels, y, color='skyblue')
plt.xlabel('Make and Model') 
plt.ylabel('Number of Accidents')
plt.title('Top Makes and Models with Most Accidents')
```

```python
#barchart showing weather conditions with most accidents

plt.figure(figsize=(12, 6))
x=weather_conditions.head(10)
y= x['Total_Accidents']
labels = x['weather_condition']
plt.bar(labels, y, color='skyblue')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.title('Weather Conditions with Most Accidents')
```

```python
#barchart for fatality count in top 10 makes
plt.figure(figsize=(12, 6))
x=fatality_analysis.head(10)    
y= x['total_fatal_injuries']
labels = x['make']
plt.bar(labels, y, color='salmon')
plt.xlabel('Make')
plt.ylabel('Total Fatal Injuries')
plt.title('Top 10 Makes with Highest Fatal Injuries')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
