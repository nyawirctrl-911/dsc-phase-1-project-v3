## DATA CLEANING

```python
used_columns = df[['Make','Model','Injury.Severity','Aircraft.damage','Aircraft.Category',
                   'Number.of.Engines','Engine.Type','Purpose.of.flight',
                   'Weather.Condition','Total.Fatal.Injuries',
                   'Total.Serious.Injuries','Total.Minor.Injuries','Registration.Number']]

#It organizes the column titles by removing spaces, converting to lowercase and replacing periods with underscores
used_columns.columns = (
    used_columns.columns
    .str.strip()
    .str.lower()
    .str.replace('.', '_', regex=False)
)
used_columns
```

```python
used_columns.isna().sum() #counts no. of null values
```

```python
used_columns=used_columns.dropna() #removes null values
used_columns
```
